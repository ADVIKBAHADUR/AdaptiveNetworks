```python
import networkx as nx
import math
import matplotlib.pyplot as plt

#Create a test graph
m=3  #Number of initial links
n=12 #Number of nodes
ncols=4 #Number of columns in a 10x10 grid of positions
G=nx.barabasi_albert_graph(n, m)
pos = nx.spring_layout(G)
position = {i : (i // ncols, (n-i-1) % ncols) for i in G.nodes()}

#Compute the node-to-node distances
lengths={}
for edge in G.edges():
   startnode=edge[0] 
   endnode=edge[1]
   lengths[edge]=round(math.sqrt(((position[endnode][1]-position[startnode][1])**2)+
                                      ((position[endnode][0]-position[startnode][0])**2)),2) #The distance
fig, axes = plt.subplots(1,1,dpi=72)
nx.draw(G, pos=pos, ax=axes, with_labels=True)
print(lengths)
```
